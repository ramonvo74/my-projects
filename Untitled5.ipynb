```python
import numpy as np
ave = 2.0
std = np.sqrt(4.0)
N = 10000
np.random.seed(1000)
X = np.random.normal(ave, std, N)
```

```python
print(X)
```

```python
ave = 2.0
std = 4.0
N = 10000
np.random.seed(0)
X = np.random.normal(ave, std, N)
print(X)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.hist(X,bins=25,normed=True)
ave = 2.0
var = 4.0
x = np.arange(-15,15,0.01)
y = np.exp(-(x-ave)**2/(2*var))/np.sqrt(2*np.pi*var)
plt.xlim(-15,15)
plt.plot(x,y,color='b',lw=4)
plt.xlabel(r'$x$',fontsize=20)
plt.ylabel(r'$P(x)$',fontsize=20)
plt.legend([r'$\langle X\rangle=2, \sigma^2=4$',r'histgram'], fontsize=16)
plt.show()
```
